# How many students are drinking alcohol illegally?
In Portugal, 17 is the legal age to drink Beer, Wine and other light drinks and 18 is the legal age to drink hard liquor. Assuming, students who have provided 1 as the input in 'Dalc' and 'Walc' columns, are the students who do not drink at all.

```python
# Converting the Pstatus (Parents Status) to binary form in 0 and 1 , so they can be averaged
port_df['Pstatus'] = port_df['Pstatus'].map({'T': 1, 'A': 0})
math_df['Pstatus'] = math_df['Pstatus'].map({'T': 1, 'A': 0})
# calculating it for the Math class-
math_alcohol_df = math_df[math_df.age < 17]
math_illegal_alcohol_df = math_alcohol_df[math_alcohol_df.Walc >1]
print('Percentage of students drinking alcohol illegally is - ',(len(math_illegal_alcohol_df)/len(math_df)) * 100, '%' )
```

```python
# calculating it for the Portugese class-
port_alcohol_df = port_df[port_df.age < 17]
port_illegal_alcohol_df = port_alcohol_df[port_alcohol_df.Walc >1]
print('Percentage of students drinking alcohol illegally is - ',(len(port_illegal_alcohol_df)/len(port_df)) * 100, '%' )
```

Both in maths and Portuguese class, the percentage of students drinking alcohol illegally, is same. It will be interesting to see if illegal drinking is related to a family situation. Like parents are living apart or the student is living with a guardian.

# What can be the reason of drinking alcohol illegally or what are the symptoms that a student is drinking alcohol illegally?

```python
# Checking for the math class
math_non_alcohol_df = math_alcohol_df[math_alcohol_df.Walc <=1]
print('Mean of different features for student drining alcohol illegally')
print(math_illegal_alcohol_df.describe().loc['mean'])
print('Mean of different features for underage students who do not drink alcohol')
print(math_non_alcohol_df.describe().loc['mean'])
```

It can be observed that there is no significant difference in the average of Pstatus. Therefore, the parent status whether they are living together or separate does not contribute to underage drinking. However, if you look at the 'goout' feature, which represents the time spent with friends outside, it is high for students who drink illegally, i.e., 3.23 as compared to the students who do not drink, i.e., 2.7.      
Also, Number of absence is greater for the student who drinks illegally, i.e., 5.51 as compared to nondrinking students, i.e., 3.4.      
Last, underage students who do not drink alcohol has strong family relationship average of 4.07 as compared to the drinking students, i.e., 3.75
