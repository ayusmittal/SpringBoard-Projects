```python
# Statistics for the CapStone Project-1
# Importing the module
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm as cm
```

```python
# Load the data sets-
math_df = pd.read_csv('student-mat.csv')
port_df = pd.read_csv('student-por.csv')
```

# what is the correlation among different variables?

```python
#Finding the correlation of each variable with the final grades Maths-

xps = math_df.corr()
cmap = cm.get_cmap('jet', 60)
plt.figure(figsize=(18,18))
plt.matshow(xps, cmap=cmap, fignum =1)
clb = plt.colorbar()
#clb.ax.set_title('This is a title')
columns = list(xps)
x_pos = np.arange(len(columns))
y_pos = np.arange(len(columns))
plt.title(' Feature Correlation')
plt.xticks(x_pos,columns)
plt.yticks(y_pos,columns)
plt.show()


```

It is observed that there is a lot of negative correlation between different variables. However, the first term and mid term grades have a high positive correlation with the final grades. Observations-    
1) Higher the mother and father education, higher the score of students.    
2) Number of failures are negatively correlated with the student final grade, it means that higher the number of failures, lower the score of a student.    
3) Going out with friends frequently and large alcohol consumptions are negatively correlated with the student final grades.    

```python
#Finding the correlation of each variable with the final grades Portuguese-

xps = port_df.corr()
cmap = cm.get_cmap('jet', 60)
plt.figure(figsize=(18,18))
plt.matshow(xps, cmap=cmap, fignum =1)
clb = plt.colorbar(cmap = cmap)
#clb.ax.set_title('This is a title')
columns = list(xps)
x_pos = np.arange(len(columns))
y_pos = np.arange(len(columns))
plt.title(' Feature Correlation')
plt.xticks(x_pos,columns)
plt.yticks(y_pos,columns)
plt.show()
```

# What is the association of final grades with the categorical features?
# 1) Does inspiration for higher education make changes in final grades?

```python
extract_df = math_df[math_df.higher == 'yes']
t1 = sum(extract_df.G3)/len(extract_df)
extract_df2 = math_df[math_df.higher == 'no']
t2 = sum(extract_df2.G3)/len(extract_df2)
print('Average score (Maths) of students who wants to go for higher education- ',t1)
print('Average score (Maths) of students who DO NOT want to go for higher education- ', t2)
pextract_df = port_df[port_df.higher == 'yes']
p1 = sum(pextract_df.G3)/len(pextract_df)
pextract_df2 = port_df[port_df.higher == 'no']
p2 = sum(pextract_df2.G3)/len(pextract_df2)
print('Average score (Portuguese) of students who wants to go for higher education- ',p1)
print('Average score (Portuguese) of students who DO NOT want to go for higher education- ', p2)
my_columns = ["Aspire for Higher Ed.", "No Higher Ed. "]
df = pd.DataFrame([[t1,t2], [p1,p2]])
df.columns = my_columns
df.index = ["Maths", "Portuguese"]
df
```

It is clearly obsereved students with an aim of higher education are serious and get better grades in final Maths test.
