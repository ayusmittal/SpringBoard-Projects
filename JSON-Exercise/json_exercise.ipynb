## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import numpy as np
## Finding the top 10 countries with the most projects
sample_json_df = pd.read_json('data/world_bank_projects.json')
s = sample_json_df.groupby('countryname').size().sort_values(ascending = False)
s = s.head(10)
print 'Top 10 countries with the most project'
s


```

```python
## Loading and Normalizing the JSON files
temp = json.load((open('data/world_bank_projects.json')))
table = json_normalize(temp,'mjtheme_namecode')

# creating a copy of table
table_copy = table
table_copy2 = json_normalize(temp,'mjtheme_namecode')

```

```python
# this is the table which contains the code and name  of the project
table['name'].replace('', np.nan, inplace=True)
table.dropna(subset=['name'], inplace=True)
table.drop_duplicates()
print 'Table having unique code and name of the project'
ValueTable = table.drop_duplicates()
ValueTable
```

```python
## Finding the top 10 projects

codes = table_copy['code'].value_counts().sort_values(ascending = False).index.tolist()
print 'The top projects are :-'
for i in codes:
    name = table.loc[table['code'] == i, 'name'].iloc[0]
    print name
```

```python
## Filling the data frame with the missing names of the project
for i in range(0,len(table_copy2)):
    if table_copy2.loc[i,'name'] == '':
        code = table_copy2.loc[i,'code']
        name = str(ValueTable.loc[ValueTable['code'] == code, 'name'].iloc[0])
        table_copy2.loc[i,'name'] = name
print 'The filled Data frame'
table_copy2
```
