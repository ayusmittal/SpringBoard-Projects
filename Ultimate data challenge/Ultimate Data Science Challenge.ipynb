# Part 2 Experiment and metrics design

The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on
weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active
during the day. On weekends, there is reasonable activity in both cities.
                           
However, a toll bridge, with a two way
toll, between the two cities causes driver partners to tend
to be exclusive to each city. The Ultimate managers of city operations for the two cities have
proposed an experiment to encourage driver partners to be available in both cities, by
reimbursing all toll costs.           

1. What would you choose as the key measure of success of this experiment in
encouraging driver partners to serve both cities, and why would you choose this metric?
                 
2. Describe a practical experiment you would design to compare the effectiveness of the
proposed change in relation to the key measure of success. Please provide details on:
      
    a. how you will implement the experiment 
    
    b. what statistical test(s) you will conduct to verify the significance of the observation     
   
    c. how you would interpret the results and provide recommendations to the city operations team along with any caveats.   
  

```python
# Solution
```

The overall objective of paying the toll fee to drivers is to increase the revenue and make the cabs available in both the cities. But, is it worth to pay toll fee? Will it generate enough revenue to the company?    

1) The metric would be two things. First, how many crossings are done by the driver and second how much bookings have increased when the driver crosses the bridge.        

2) A practical experiment would be A/B testing in this scenario.    

a) In the first month, when the company is not  paying toll fee, following things need to be measured - 

1) measure how much requests are generated in both cities    
2) is there any waiting time for the customers when they request a cab
3) How much is the total net profit genrated in both cities.    
    
In the second month, ignore any additional drivers who joined, provide the promo offer to the existing drivers. And measure following things-    
     
1) How many crossings made, and how much money paid to drivers as reimbusement.    
2) Is there change in waiting time for the customers, since the more cabs are available now.     
3) How much net profit is generated in the present month with the promo offer.    
    
Once We have all this information, we can compare if the net profit, increased, decreased or it was constant. If there is less 
waiting time during the second month, it means the user experience is enhanced and likely in future more people will request rides because of low waiting time. finaly if the net profit increase, the hypotheses if reimbursing toll fee is validated.
Hence, reimbursing the toll fee is in the favor of company.

```python
# Implementing a random Forest classifer to see which features are the most importance.
df6.head(5)
```

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
df6.columns
```

Removing the dates from the data set because they will not play a critical role in the decision. Also, city and phone are 
categorical variables, need to change them to the one hot enconding.

```python
mldf = pd.get_dummies(df6, columns=["city", "phone", "ultimate_black_user"])
mldf['target'] = mldf['trips_in_first_30_days'] >0
mldf = mldf.drop(['last_trip_date','signup_date','trips_in_first_30_days'], axis = 1)
# Splitting the data for training and testing
target = mldf.target
mldf = mldf.drop('target', axis =1 )
train = mldf.values
eng_train, eng_test, eng_target_train, eng_target_test = train_test_split(train, target, test_size=0.10, random_state=42)
# Training the model
rfc = RandomForestClassifier(n_jobs=-1, random_state=0, n_estimators =600, max_depth = 15)
rfc.fit(eng_train, eng_target_train)
# Calculating accuracy on the testing model
scores = accuracy_score(eng_target_test, rfc.predict(eng_test))
print(' The accuracy is ', scores * 100, '%')
```

The implemented predicted model has given an accuracy of ~72 % on the testing data.

```python
important_features = rfc.feature_importances_
important_features
```

```python
columns = mldf.columns
print('The most important features are -')
i =0
for each in columns:
    print('Feature name ', each, ', importance - ',important_features[i] )
    i =i+1
```
